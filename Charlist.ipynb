```python
def number_to_char(n):
    n = n % 26
    return chr(n+65) #0 = A, 1 = B, ..., 24 = Y, 25 = Z
    
def charlen_from_str(start,end):
    startsum = 0
    for i in range(len(start)):
        startsum += (ord(start[-(i+1)]) - 64)*26**i
        
    endsum = 0
    for i in range(len(end)):
        endsum += (ord(end[-(i+1)]) - 64)*26**i

    return endsum - startsum


def make_charlist(start,end):
    charlen = charlen_from_str(start,end) + 1
    starti = charlen_from_str("A",start) + 1
    
    charlist = []
    for i in range(charlen):
        nextchar = ""
        num = int((i+starti-1)/26) 
        if num > 0:
            nextchar += number_to_char(num-1) 
        nextchar += number_to_char(i) 
        charlist.append(nextchar)
    return charlist

charlist = make_charlist("A","CZ")
for char in charlist:
    print(char)
```

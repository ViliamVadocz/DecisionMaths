```python
from math import cos, sin, atan2
import numpy as np
import matplotlib.pyplot as plt

#A rocket produces a constant acceleration due to it's propulsion in the direction it is traveling
#What curve does it follow?

def simulation_tick(pos,vel,ang,grav,prop):
    new_pos = pos + vel
    new_vel = vel + grav + prop
    new_ang = atan2(new_vel[1],new_vel[0])
    new_prop = np.array([cos(new_ang),sin(new_ang)])
    return new_pos, new_vel, new_ang, new_prop

#gravity is 1 unit/tick downwards
#propulsion is 1 unit/tick in direction of velocity
#initial velocity is 100 units/tick in direction of angle
#angle is pi / 4

pos  = np.zeros(2)
ang  = np.pi / 4
vel  = 100 * np.array([cos(ang),sin(ang)])
grav = np.array([0,-1])
prop = np.array([cos(ang),sin(ang)])

n = 200
s = np.zeros((n,2))

for i in range(n):
    s[i] += pos
    pos, vel, ang, prop = simulation_tick(pos,vel,ang,grav,prop)

#Plotting
plt.subplots()
plt.plot(s[:,0], s[:,1], 'b-', label='Curve')
plt.title('Rocket Curve Problem')
plt.xlabel('x')
plt.ylabel('y')
plt.grid()
plt.show()
```
